````python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
import pandas as pd

# Esempio di DataFrame
df1 = countries
df2 = currency
````

````python
# Funzione per calcolare la similarità tra due stringhe
def calculate_similarity(s1, s2):
    return fuzz.token_sort_ratio(s1, s2)  # Utilizziamo token_sort_ratio per una maggiore flessibilità

# Creazione di un dizionario per memorizzare le corrispondenze tra i valori non univoci di 'countries' in df1 e i corrispondenti valori in df2
matches = {}

# Iterazione sui valori non univoci della colonna 'countries' in df1
for country1 in df1['countries'].unique():
    max_similarity = 0
    best_match = None
    # Iterazione sui valori della colonna 'countries' in df2
    for country2 in df2['countries']:
        # Calcolo dell'indice di similarità tra le stringhe
        similarity = calculate_similarity(country1, country2)
        # Aggiornamento della migliore corrispondenza
        if similarity > max_similarity:
            max_similarity = similarity
            best_match = country2
    # Aggiungiamo la migliore corrispondenza al dizionario
    matches[country1] = best_match

# Creazione di una nuova colonna nel DataFrame df1 con i valori corrispondenti trovati in df2
df1['matched_country'] = df1['countries'].map(matches)
````

````python
df1=pd.merge(df1,df2,left_on='matched_country',right_on='countries')
````

````python
# Funzione per calcolare i pesi normalizzati delle valute e dei paesi per tutti i ticker nel portafoglio
def calculate_normalized_weights(config, countries):
    currency_weights = {}
    country_weights = {}

    for etf_info in config['etf']:
        ticker_symbol = etf_info[0]
        ticker_countries = countries[countries['symbol'] == ticker_symbol]

        # Rimuovi spazi bianchi alla fine della stringa nella colonna 'percentage' e converti i valori in numeri decimali
        ticker_countries['percentage'] = pd.to_numeric(ticker_countries['percentage'].str.strip()) / 100

        # Calcola i pesi normalizzati delle valute per il ticker
        currency_weights[ticker_symbol] = ticker_countries.groupby('currency')['percentage'].sum()

        # Calcola i pesi normalizzati dei paesi per il ticker
        country_weights[ticker_symbol] = ticker_countries.groupby('countries')['percentage'].sum()

    return currency_weights, country_weights

# Calcola i pesi normalizzati delle valute e dei paesi
currency_weights, country_weights = calculate_normalized_weights(config, countries)

# Salva i pesi normalizzati delle valute in un DataFrame
df_currency_weights = pd.DataFrame(currency_weights)
# Salva i pesi normalizzati dei paesi in un DataFrame
df_country_weights = pd.DataFrame(country_weights)
````

````python
from yahooquery import Ticker

# Funzione per calcolare i pesi delle valute per ogni ticker
def calculate_currency_weights(config, df_currency_weights):
    for ticker in df_currency_weights.columns:
        try:
            # Otteniamo il nome completo del ticker
            long_name = Ticker(ticker).quotes[ticker]['longName']

            # Verifichiamo se la parola "Hedged" è presente nel nome del ticker e se la valuta è "EUR"
            if "Hedged" in long_name and "EUR" in long_name:
                # Impostiamo il peso della valuta "EUR" al 100% per questo ticker
                df_currency_weights[ticker] = 0
                df_currency_weights.at["EUR", ticker] = 1
        except KeyError:
            # Se non possiamo ottenere il nome completo del ticker, passiamo al prossimo ticker
            pass

    return df_currency_weights

# Modifica della tabella df_currency_weights
df_currency_weights_modified = calculate_currency_weights(config, df_currency_weights)
````

````python
# 2. Calcolare il patrimonio totale del portafoglio
# Ottenere il prezzo attuale di ogni titolo dal ticker e calcolare il controvalore attuale
````

````python
def calculate_current_portfolio_value(df, t):
    current_portfolio_value = 0
    for ticker in df['Ticker'].unique():
        current_price = Ticker(ticker).quotes[ticker]['regularMarketPrice']
        quantity = df[df['Ticker'] == ticker]['Quantita'].sum()
        current_portfolio_value += current_price * quantity
    return current_portfolio_value

# Calcolare il valore attuale del portafoglio
current_portfolio_value = calculate_current_portfolio_value(df, t)
print("Valore attuale del portafoglio:", current_portfolio_value)
````

````python
# 3. Calcolare il rendimento totale e percentuale del portafoglio
def calculate_total_return(initial_value, current_value):
    total_return = current_value - initial_value
    total_return_percentage = (total_return / initial_value) * 100
    return total_return, total_return_percentage

# Calcolare il rendimento totale e percentuale del portafoglio
initial_portfolio_value = (df['Controvalore']).sum() 
total_return, total_return_percentage = calculate_total_return(initial_portfolio_value, current_portfolio_value)

print("Rendimento totale del portafoglio:", total_return)
print("Rendimento percentuale del portafoglio:", total_return_percentage, "%")
````

````python
import pandas as pd

# Inizializza una lista vuota per contenere i dizionari delle righe del DataFrame
portfolio_data = []

# Itera su ogni ticker nel DataFrame
for ticker in df['Ticker'].unique():
    # Ottieni il TER dal DataFrame df
    ter = df[df['Ticker'] == ticker]['TER'].iloc[0]
    
    # Ottieni il prezzo attuale del ticker
    current_price = Ticker(ticker).price[ticker]['regularMarketPrice']
    
    # Ottieni le quantità possedute per il ticker
    quantities = df[df['Ticker'] == ticker]['Quantita']
    
    # Calcola il controvalore moltiplicando la quantità posseduta per il valore attuale del titolo
    controvalore = quantities * current_price
    
    # Aggiungi una riga alla lista portfolio_data con le informazioni del ticker
    portfolio_data.append({
        'Ticker': ticker,
        'Ter': ter,
        'Quantita': quantities.sum(),  # Somma delle quantità possedute per il ticker
        'Valore Attuale': current_price,
        'Controvalore': controvalore.sum()
    })

# Crea il DataFrame dal list di dizionari
portfolio_current = pd.DataFrame(portfolio_data)
````

````python
bond_list=[]

from yahooquery import Ticker
import pandas as pd

def get_sector_weightings(portfolio_current):
    # Extract ticker symbols from the 'Ticker' column of portfolio_current
    ticker_symbols = portfolio_current['Ticker'].tolist()

    # Initialize an empty list to store sector weightings DataFrames
    sector_weightings_list = []

    # Iterate over each ticker symbol
    for symbol in ticker_symbols:
        # Fetch sector weightings for the current ticker
        t = Ticker(symbol)
        category_holdings_df = t.fund_category_holdings
        
        # Check if the DataFrame contains 'bondPosition' column
        if 'bondPosition' in category_holdings_df.columns:
            # Check if 'bondPosition' column has a value greater than 0.5
            if category_holdings_df['bondPosition'].iloc[0] > 0.5:
                print(f"Bond position is greater than 0.5 for {symbol}. Skipping...")
                bond_list.append(symbol)
                continue

        sector_weightings_df = t.fund_sector_weightings

        # Check if sector_weightings_df contains data
        if not sector_weightings_df.empty:
            # Extract the ticker symbol from the DataFrame
            ticker_symbol = sector_weightings_df.columns[0]

            # Reset the index to make 'Sector' a column
            sector_weightings_df.reset_index(inplace=True)

            # Rename the index column to 'Sector'
            sector_weightings_df.rename(columns={'index': 'Sector'}, inplace=True)

            # Print sector_weightings_df and its column names for debugging
            
            # Add the ticker's sector weightings to the list
            sector_weightings_list.append(sector_weightings_df)
        else:
            print(f"No sector weightings available for {symbol}. Skipping...")

    # Check if there are sector weightings available for any ticker
    if not sector_weightings_list:
        return None

    # Concatenate sector weightings for all tickers
    merged_sector_weightings = pd.concat(sector_weightings_list, axis=0)

    # Print the merged DataFrame and its column names for debugging

    return merged_sector_weightings

# Example usage:
# sector_weightings = get_sector_weightings(portfolio_current)
# print(sector_weightings)
````

````python
bond_list=[]

from yahooquery import Ticker
import pandas as pd

def get_holdings_weightings(portfolio_current):
    # Extract ticker symbols from the 'Ticker' column of portfolio_current
    ticker_symbols = portfolio_current['Ticker'].tolist()

    # Initialize an empty list to store sector weightings DataFrames
    holdings_weightings_list = []

    # Iterate over each ticker symbol
    for symbol in ticker_symbols:
        # Fetch sector weightings for the current ticker
        t = Ticker(symbol)
        category_holdings_df = t.fund_category_holdings
        
        # Check if the DataFrame contains 'bondPosition' column
        if 'bondPosition' in category_holdings_df.columns:
            # Check if 'bondPosition' column has a value greater than 0.5
            if category_holdings_df['bondPosition'].iloc[0] > 0.5:
                print(f"Bond position is greater than 0.5 for {symbol}. Skipping...")
                bond_list.append(symbol)
                continue

        
        top_holdings_df=t.fund_top_holdings

        # Check if sector_weightings_df contains data
        if not top_holdings_df.empty:
            
            top_holdings_df = top_holdings_df.rename(columns={'symbol':'Ticker'})

            top_holdings_df = top_holdings_df.reset_index()

            top_holdings_df.rename(columns={'symbol': 'new_name'}, inplace=True)

            top_holdings_df = top_holdings_df[['Ticker','holdingName','holdingPercent']]

            top_holdings_df = top_holdings_df.rename(columns={'holdingPercent':symbol})
            # Add the ticker's sector weightings to the list
            holdings_weightings_list.append(top_holdings_df)
            
        else:
            print(f"No sector weightings available for {symbol}. Skipping...")

    # Check if there are sector weightings available for any ticker
    if not holdings_weightings_list:
        return None

    # Concatenate sector weightings for all tickers
    merged_holdings_weightings = pd.concat(holdings_weightings_list, axis=0)

    # Print the merged DataFrame and its column names for debugging

    return merged_holdings_weightings

# Example usage:
# sector_weightings = get_sector_weightings(portfolio_current)
# print(sector_weightings)
````

````python
# Plotto le linee di MWRR e TWRR
````

````python
import yfinance as yahooFinance
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assicurati che le date nel dataframe df siano senza fuso orario (naive)
df['Data valuta'] = pd.to_datetime(df['Data valuta']).dt.tz_localize(None)

# Lista dei ticker unici nel tuo portafoglio
tickers = df['Ticker'].unique()

# Creare un dizionario per immagazzinare i dati storici per ogni ticker
ticker_history = {}

# Estrarre i dati storici dei prezzi per ogni ticker
for ticker in tickers:
    timeseriesticker = yahooFinance.Ticker(ticker)
    data = timeseriesticker.history(period="max").reset_index()
    # Assicurati che le date siano senza fuso orario (naive)
    data['Date'] = data['Date'].dt.tz_localize(None)
    ticker_history[ticker] = data[['Date', 'Close']]

# Trovare il primo e l'ultimo giorno nel portafoglio
first_date = df['Data valuta'].min()
last_date = pd.Timestamp.today()

# Creare un intervallo di date dal primo giorno fino ad oggi
all_dates = pd.date_range(start=first_date, end=last_date, freq='D')

# Creare un dataframe vuoto per immagazzinare le quantità e i valori posseduti giornalmente
quantity_over_time = pd.DataFrame(all_dates, columns=['Date'])

# Creare un dataframe vuoto per immagazzinare il valore del portafoglio giornaliero
value_over_time = pd.DataFrame(all_dates, columns=['Date'])

# Dizionario per immagazzinare le quantità iniziali per ciascun ticker
initial_quantities = {}

# Iterare su ciascun ticker per calcolare le quantità possedute giornaliere
for ticker in tickers:
    # Filtrare le operazioni relative al ticker corrente
    operations = df[df['Ticker'] == ticker]

    # Creare una serie temporale con le date e le quantità
    quantities = pd.Series(index=all_dates, data=0)
    
    for index, row in operations.iterrows():
        # Data della transazione e quantità
        trans_date = row['Data valuta']
        quantity = row['Quantita']
        
        # Aggiungere le quantità dalla data della transazione in poi
        quantities.loc[trans_date:] += quantity

    # Salvare la quantità iniziale per il ticker
    initial_quantities[ticker] = quantities.iloc[0]

    # Aggiungere la serie delle quantità al dataframe delle quantità
    quantity_over_time[ticker] = quantities.values

    # Unire le quantità con i prezzi di chiusura
    price_history = ticker_history[ticker].set_index('Date').reindex(all_dates).ffill()
    value_history = quantities * price_history['Close']

    # Aggiungere la serie dei valori al dataframe dei valori
    value_over_time[ticker] = value_history.values

# Calcolare il valore totale del portafoglio giornaliero sommando i valori dei singoli ticker
value_over_time['Total Value'] = value_over_time[tickers].sum(axis=1)

# Calcolare i versamenti cumulati nel tempo
df['Investment'] = df['Quantita'] * df['Prezzo']
investment_over_time = pd.Series(index=all_dates, data=0)
for index, row in df.iterrows():
    trans_date = row['Data valuta']
    investment = row['Investment']
    investment_over_time.loc[trans_date:] += investment

value_over_time['Cumulative Investment'] = investment_over_time.values

# Calcolare la differenza tra Cumulative Investment e Total Value
value_over_time['Difference'] = value_over_time['Total Value'] - value_over_time['Cumulative Investment']

# Calcolare il rendimento della colonna calcolata (MWRR)
value_over_time['MWRR'] = value_over_time['Difference'] / value_over_time['Cumulative Investment']

# Gestire i casi in cui Cumulative Investment è 0 per evitare divisioni per 0
value_over_time['MWRR'] = value_over_time['MWRR'].replace([np.inf, -np.inf], np.nan).fillna(0)

# Calcolare il TWRR (Time-Weighted Rate of Return)
rows = []
for date in pd.date_range(start=first_date, end=last_date, freq='D'):
    row_data = {'Date': date}
    for ticker in tickers:
        # Trovare il prezzo del ticker per la data corrente
        price = ticker_history[ticker][ticker_history[ticker]['Date'] == date]['Close'].values
        if len(price) > 0:
            price = price[0]
        else:
            # Prezzo non disponibile per la data corrente, riempi con il prezzo del giorno precedente
            previous_price = ticker_history[ticker][ticker_history[ticker]['Date'] < date]['Close'].iloc[-1]
            price = previous_price if not np.isnan(previous_price) else 0
        # Calcolare il valore del portafoglio giornaliero per il ticker corrente
        value = initial_quantities[ticker] * price
        row_data[ticker] = value
    rows.append(row_data)

# Creare un dataframe dai dati raccolti
TWRR_data = pd.DataFrame(rows)

# Calcolare il valore totale del portafoglio giornaliero sommando i valori dei singoli ticker
TWRR_data['Total Value'] = TWRR_data[tickers].sum(axis=1)

initial_portfolio_value = value_over_time.iloc[0]['Cumulative Investment']

# Calcolare la differenza tra Total Value e Valore Portafoglio iniziale
TWRR_data['Difference'] = TWRR_data['Total Value'] - initial_portfolio_value

# Calcolare il rendimento del portafoglio giornaliero (TWRR)
TWRR_data['TWRR'] = TWRR_data['Difference'] / initial_portfolio_value

# Gestire i casi in cui Valore Portafoglio iniziale è 0 per evitare divisioni per 0
TWRR_data['TWRR'] = TWRR_data['TWRR'].replace([np.inf, -np.inf], np.nan).fillna(0)
````

````python
# Tracciare il trend del MWRR e del TWRR nel tempo
plt.figure(figsize=(14, 7))
# Plotting MWRR
plt.plot(value_over_time['Date'], value_over_time['MWRR'], label='MWRR', color='blue')

# Plotting TWRR
plt.plot(TWRR_data['Date'], TWRR_data['TWRR'], label='TWRR', color='red')

# Adding labels and title
plt.xlabel('Data')
plt.ylabel('Rendimento')
plt.title('Confronto tra MWRR e TWRR nel Tempo')
plt.legend()
plt.grid(True)
plt.show()
````

````python
import yfinance as yahooFinance
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assicurati che le date nel dataframe df siano senza fuso orario (naive)
df['Data valuta'] = pd.to_datetime(df['Data valuta']).dt.tz_localize(None)

# Lista dei ticker unici nel tuo portafoglio
tickers = df['Ticker'].unique()

# Creare un dizionario per immagazzinare i dati storici per ogni ticker
ticker_history = {}

# Estrarre i dati storici dei prezzi per ogni ticker tramite Yahoo Finance
for ticker in tickers:
    timeseriesticker = yahooFinance.Ticker(ticker)
    data = timeseriesticker.history(period="max").reset_index()
    # Assicurati che le date siano senza fuso orario (naive)
    data['Date'] = pd.to_datetime(data['Date']).dt.tz_localize(None)
    ticker_history[ticker] = data[['Date', 'Close']]

# Trovare il primo e l'ultimo giorno di ogni anno nella storia del portafoglio
first_date = df['Data valuta'].min()
last_date = pd.Timestamp.today()

# Creare un intervallo di date dal primo giorno fino ad oggi
all_dates = pd.date_range(start=first_date, end=last_date, freq='D')

# Creare un dataframe vuoto per immagazzinare le quantità e i valori posseduti giornalmente
quantity_over_time = pd.DataFrame(all_dates, columns=['Date'])

# Creare un dataframe vuoto per immagazzinare il valore del portafoglio giornaliero
value_over_time = pd.DataFrame(all_dates, columns=['Date'])

# Creare i versamenti cumulati nel tempo
df['Investment'] = df['Quantita'] * df['Prezzo']
investment_over_time = pd.Series(index=all_dates, data=0)
for index, row in df.iterrows():
    trans_date = row['Data valuta']
    investment = row['Investment']
    investment_over_time.loc[trans_date:] += investment

# Iterare su ciascun ticker per calcolare le quantità possedute giornalmente
for ticker in tickers:
    # Filtrare le operazioni relative al ticker corrente
    operations = df[df['Ticker'] == ticker]

    # Creare una serie temporale con le date e le quantità
    quantities = pd.Series(index=all_dates, data=0)
    
    for index, row in operations.iterrows():
        # Data della transazione e quantità
        trans_date = row['Data valuta']
        quantity = row['Quantita']
        
        # Aggiungere le quantità dalla data della transazione in poi
        quantities.loc[trans_date:] += quantity

    # Aggiungere la serie delle quantità al dataframe delle quantità
    quantity_over_time[ticker] = quantities.values

    # Unire le quantità con i prezzi di chiusura
    price_history = ticker_history[ticker].set_index('Date').reindex(all_dates).fillna(method='ffill')
    value_history = quantities * price_history['Close']

    # Aggiungere la serie dei valori al dataframe dei valori
    value_over_time[ticker] = value_history.values

# Calcolare il valore totale del portafoglio giornaliero sommando i valori dei singoli ticker
value_over_time['Total Value'] = value_over_time[tickers].sum(axis=1)

# Calcolare i rendimenti giornalieri del portafoglio
value_over_time['Daily Return'] = value_over_time['Total Value'].pct_change()

# Aggiungere una colonna per l'anno
value_over_time['Year'] = value_over_time['Date'].dt.year

# Calcolare la volatilità annuale come la deviazione standard dei rendimenti giornalieri annualizzati
annual_volatility = value_over_time.groupby('Year')['Daily Return'].agg(['std', 'count'])
annual_volatility['Annualized Volatility'] = annual_volatility['std'] * np.sqrt(annual_volatility['count'])

# Tracciare la volatilità annuale in un grafico a barre
plt.figure(figsize=(10, 6))
plt.bar(annual_volatility.index.astype(str), annual_volatility['Annualized Volatility'], color='skyblue')
plt.title('Volatilità Annuale del Portafoglio')
plt.xlabel('Anno')
plt.ylabel('Volatilità Annuale')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
````

Calcolare le performance degli investimenti: un errore comune
Domanda semplice, risposta non scontata: come si calcola la performance di un investimento? Spesso si fa l’errore di valutare solo il patrimonio di partenza e il saldo finale. Ad esempio: investiamo 1.000 € a inizio marzo e otteniamo a fine mese 1.100 € (+10%). Il mese successivo decidiamo di investire altri 9.000 €, ma questa volta il risultato è -10%. A fine aprile avremo perso in tutto 900 € su 10.000 € investiti. Analizzando questo esempio, si potrebbe essere delusi dal proprio investimento e pensare che il rendimento sul versato sia di -9%. In realtà, il calcolo corretto dovrebbe coinvolgere diversi fattori: importo, movimenti effettuati (per rimpinguare o ridurre la cifra investita), tempistiche e le caratteristiche del Piano di Accumulo Capitale. Per includere il più possibile le diverse variabili che partecipano alla performance, si utilizzano il Money Weighted Rate of Return (MWRR) e il Time Weigthed Rate of Return (TWRR).

Cos’è il MWRR: una definizione
Il MWRR può essere definito come il tasso di rendimento ponderato in base al capitale effettivamente investito. Tiene quindi conto non solo della cifra iniziale e del saldo finale ma anche dei flussi e dell’ampiezza del periodo d’investimento.  

Per tornare all’esempio iniziale, il MWRR fa luce sul percorso che intercorre tra patrimonio di partenza e il saldo. Quindi, prende in considerazione il conferimento effettuato ad aprile e l’andamento degli investimenti nel tempo. Il risultato di gestione è pari a -900 € (somma di cifra iniziale, saldo, conferimenti e riscatti). Si tratta però di una cifra assoluta che poco dice sulla performance. Il rendimento Money Weighted sarà di -15% perché dato dal rapporto tra il risultato di gestione -900 € e il patrimonio medio che è il risultato di (1.050 € + 9.500 €) / 2. L’indicatore, che come si nota offre un dato diverso da quello semplificato (ottenuto dal rapporto tra investimento iniziale e saldo) è utile per valutare il rendimento effettivo delle somme mediamente investite. Attraverso il MWRR non si evidenzia però come si è comportato il gestore perché è un risultato condizionato dalle scelte e dalle strategie di investimento (nuovi conferimenti, prelievi, PAC) attivate durante il periodo. 

Come funziona il TWRR
Le performance MWRR di un portafoglio non rappresentano la capacità del gestore di remunerare il capitale investito. Non è infatti il gestore a decidere i flussi di ingresso e uscita e non è quindi giusto attribuirgli un peso diverso per una prima parte andata magari bene con poco capitale e una andata meno bene con molto capitale. Per far emergere questo aspetto, esiste un’altra modalità di calcolo, il Time Weigthed Rate of Return (TWRR), che sterilizza l’effetto dei flussi frazionando la performance in periodi di riferimento. Nel nostro caso, concatenando i rendimenti, il TWRR è 0% in quanto a marzo è stato ottenuto il +10% e ad aprile il -10%.

Come e quando utilizzare il rendimento TWRR e MWRR
Come possono essere utilizzati il TWRR e il MWRR per calcolare le performance degli investimenti? Sintetizzando, il rendimento TWRR può essere utile per valutare il gestore e confrontare diversi servizi di investimento (come fondi, altri gestori e Roboadvisor, ecc.).

Il rendimento MWRR, invece, può essere utile se viene confrontato con il TWRR, per capire quanto è stata efficace la strategia di conferimenti, prelievi, PAC, cambio Profilo nel tempo. Se il MWRR è maggiore del TWRR significa che le strategie sono state efficaci (magari sono state fatte sui ribassi), mentre se ciò non avviene significa che le azioni compiute hanno “inseguito” il mercato (magari delle azioni sono state comprate sui rialzi).

Ma vediamo un altro esempio. Un prodotto ha un rendimento TWRR del 7% in un anno. Marco e Chiara decidono di investire nel prodotto con 1.000 € iniziali e in seguito, di investire altri 600 € ma seguendo due strategie differenti:

Marco decide di aumentare l’investimento progressivamente, conferendo 100 € dopo ogni mese in cui le performance risultavano negative.
Chiara decide a novembre di aggiungere un conferimento di 600 € tutti in una volta.
Marco a fine periodo avrà 1.703 €, con un MWRR del 7,6% dimostrando così che la strategia attuata è stata vincente perché è riuscita ad avere una buona capacità di controllare i flussi (maggiore del 7%). Chiara avrà 1.675 €, con MWRR del 6,5%, per minore capacità.
